```python
import utilities
import surface
import pyjulip
```

```python
from ase.io import read, write
```

```python
E_surfs = []
model_i = []

for i in range(1,11):
    calculator = pyjulip.ACE("../exampleTB_com_surf/Ti_NRLTB_surf_3_N3_16_DB{}.json".format(i))
    surf = read("hcp_surf.xyz")
    bulk = read("hcp.xyz")

    bulk.set_calculator(calculator)
    surf.set_calculator(calculator)
    try:
        properties = surface.do_symmetric_surface(bulk, surf, calculator)
        E_surfs.append(properties["Ef"])
    except:
        E_surfs.append(0)
    model_i.append(i)
```

```python
import matplotlib.pyplot as plt
```

```python
E_surfs
```

```python
E_surfs[1] = 0.0

plt.plot(model_i, E_surfs, label="2B+ACE HMD")
plt.axhline(y=0.13350719987790835, color="black", label="NRLTB")
plt.xlabel("HMD iteration")
plt.ylabel("Surface formation energy [eV/Å²]")
plt.legend()
plt.ylim(0.10,0.14)
plt.tight_layout()

plt.savefig("HMD_Ti_NRLTB_surf.pdf")
```
